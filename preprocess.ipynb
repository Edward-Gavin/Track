## 根据给出的静态的货柜车数据，对货柜车进行分类

```python
# 分类器
# 根据货柜车的类型，对所有的货柜车进行汇总
def class_track(static_track_path):
    '''
        static_track_path: 静态货柜车数据文件
    '''
    data2 = pd.read_excel(static_track_path, delimiter=":")
    abc = data2.groupby("车辆经营范围")
    # print(type(abc))
    # print(abc)
    for i,j in enumerate(list(abc)):
        j[1].to_excel("./classic/"+str(i)+".xlsx")
    print("class finished")

class_track("./static.xlsx")
```
